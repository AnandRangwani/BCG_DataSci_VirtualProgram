```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_data=pd.read_csv('ml_case_training_data.csv')
raw_data.head()
```

```python
raw_data.info()
```

```python
# List of all columns that have more than 50% missing values.

array=[]
lst=list(raw_data.columns)
for i in lst:
    if raw_data[i].isnull().mean()>0.50:
        array.append(i)

array
```

```python
raw_data.drop(array,axis=1,inplace=True)
```

```python
# we also have to remove those two id's from raw_data DataFrame.
# Those two id's have index number 9687 and 11386 in raw_data.
raw_data.drop([9687,11386],axis=0,inplace=True)
```

```python
# removing those two id's from raw_data2.
raw_data2=raw_data2.loc[raw_data2.id!='8d7da64122bba29af851e7252ae36484',:]
raw_data2=raw_data2.loc[raw_data2.id!='a375227dbd198e4702f2c522c2d78f68',:]
raw_data2.reset_index(inplace=True)
raw_data2.drop('index',axis=1,inplace=True)
raw_data2.tail()
```

```python
raw_data2[raw_data2.price_p1_var.isnull()]['id'].value_counts()

# We can now see that no id's are such for which there are more than 6 missing values.
```

```python
raw_data2[raw_data2.price_p1_var.isnull()].isnull().sum()
# This shows that id's for which 'price_p1_var' is null are same for all other 'price...' features.
# Thus now we can drop all rows which are having any null value.
```

```python
raw_data2.dropna(how='any',inplace=True)
raw_data2.isnull().sum()
```

```python
raw_data2.reset_index(inplace=True)
raw_data2.drop('index',axis=1,inplace=True)
raw_data2.head()
```

```python
raw_data2.id.value_counts()
# We can see that all id's are having atleast 7 values of 'price...' features
```

```python
raw_data.id.value_counts()
# All rows in raw_data are having unique id's.
```

```python
set(raw_data.id.unique())==set(raw_data2.id.unique())
```

```python
raw_data2.describe()
```

```python
hist_data=raw_data2.groupby(['id']).mean().reset_index()
hist_data.head()
# grouping raw_data2 by id's and taking mean of all 'price...' features having same id's.
```

```python
hist_data.info()
```

```python
# joining raw_data and hist_data.
df_features=raw_data.join(other=hist_data.set_index('id'),on='id')
df_features.head()
```

```python
df_features.info()
```

```python
churn_data=pd.read_csv('ml_case_training_output.csv')
churn_data.head()
```

```python
churn_data.info()
```

```python
churn_data.churn.value_counts()
```

```python
# to remove those two id's (for which all 'price...' features were null for all 12 months of 2015) from churn_data. 
removal=[i for i in list(churn_data.id.unique()) if i not in list(df_features.id.unique())]
```

```python
removal
```

```python
churn_data=churn_data.loc[churn_data.id!=removal[0],:]
churn_data=churn_data.loc[churn_data.id!=removal[1],:]
churn_data.head(10)
```

```python
churn_data.reset_index(inplace=True)
churn_data.drop('index',axis=1,inplace=True)
churn_data.tail()
```

```python
churn_data.id.value_counts()
```

```python
# joining churn_data with df_features.
df_training=df_features.join(other=churn_data.set_index('id'),on='id')
df_training.head()
```

```python
df_training.to_csv('merged_data_training.csv')
```
